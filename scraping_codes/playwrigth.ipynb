```python
import asyncio
from playwright.async_api import async_playwright
from dotenv import load_dotenv
import os
import pandas as pd

# Load environment variables from the .env file
load_dotenv()

# Access the username and password from the environment variables
username = os.getenv('USERNAME')
password = os.getenv('PASSWORD')

# Function to scrape the recommendations
async def scrape_recommendations():
    async with async_playwright() as p:
        # Launch the browser in headful mode (GUI) to debug if headless fails
        browser = await p.chromium.launch(headless=False)  # Use headless=False to see the browser window
        context = await browser.new_context(
            user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36",
            viewport={"width": 1280, "height": 800}
        )
        
        # Open a new page
        page = await context.new_page()

        # Go to the login page
        await page.goto("https://www.marketscreener.com/login/")
        
        # Handle cookie consent if it appears
        try:
            await page.wait_for_selector('button.button__acceptAll', timeout=10000)  # Wait for the "Accept All" button
            await page.click('button.button__acceptAll')  # Click the "Accept All" button
            print("Cookie consent accepted.")
        except:
            print("No cookie consent dialog found.")
        
        # Wait for the username field to appear (extend timeout if needed)
        await page.wait_for_selector('input[name="login"]', timeout=15000)  # 15 seconds timeout
        await page.fill('input[name="login"]', username)

        # Wait for the password field to appear and fill it
        await page.wait_for_selector('input[name="password"]', timeout=15000)
        await page.fill('input[name="password"]', password)

        # Submit the form by clicking the login button
        await page.click('button[type="submit"]')

        # Wait for login to complete and check if logout option appears
        await page.wait_for_timeout(5000)  # Wait 5 seconds to ensure login
        if await page.is_visible("text=Logout"):
            print("Login successful!")
        else:
            print("Login failed, check your credentials.")
            await browser.close()
            return
        
        # Close browser after login success
        await browser.close()

# Run the Playwright function
await scrape_recommendations()
```
