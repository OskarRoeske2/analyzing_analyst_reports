```python
from fake_useragent import UserAgent
from finnews.client import News
from pprint import pprint

# Create a user-agent instance
ua = UserAgent()
headers = {'User-Agent': ua.chrome}

# Create a new instance of the News Client.
news_client = News()

# Grab the Yahoo Finance News Client.
yahoo_finance_client = news_client.yahoo_finance

# Grab the News Feed.
content = yahoo_finance_client.news(headers=headers)
pprint(content)

# Grab the headlines for Microsoft (MSFT) and Google (GOOG)
content = yahoo_finance_client.headlines(symbols=['MSFT', 'GOOG'], headers=headers)
pprint(content)
```
